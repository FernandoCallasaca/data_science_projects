```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
pd.options.display.float_format = '{:,.1f}'.format
```

```python
def path_file(name):
  return '/content/drive/My Drive/Data Science/Platzi/10. Curso de Manipulación y Análisis de datos con Pandas y Python/db/population-worldbank/{0}'.format(name)
```

```python
# https://data.worldbank.org/indicator/SP.POP.TOTL?end=2020&start=1960&view=chart
path_read_file = path_file('API_SP.POP.TOTL_DS2_en_csv_v2_3469297.csv')
path_read_file
```

```python
# aquí nuestro csv tiene 4 primera filas en blanco y luego ya inicia nuestro dataset
# es por eso eso que omitimos las 4 primeras filas utilizando skiprows
df_pob = pd.read_csv(path_read_file, skiprows = 4)
df_pob.head()
```

```python
# vamos a hacer una copia a profundidad de nuestro dataframe original o otro 
# para que ahí nomalicemos los datos como en el ejemplo
df = df_pob.copy(deep = True)
df.head()
```

```python
# primero eliminamos las columnas que no usa en el ejemplo
df.drop(columns=['Country Code','Indicator Name','Indicator Code','Unnamed: 65'], inplace = True)

# luego creamos un df tal cual el ejemplo
df = pd.melt(df, id_vars=['Country Name'], var_name='year', value_name='pob')

df.head()
```

```python
# primero cambiamos el año a una variable categórica
df[['year']] = df[['year']].astype('category')
```

```python
df.dtypes
```

```python
df.convert_dtypes().dtypes
```

```python
# cambiamos el nombre de la columna 
df.rename(columns = {'Country Name': 'Country'}, inplace = True)
df.head()
```

```python
# creamos una muestra
# pero para eso creamos un vector booleano
idx_filtro = df['Country'].isin(['Colombia', 'Peru'])
idx_filtro.sample(5)
```

```python
# Ahora podemoa hacer el filtro de nuestro df
# dentro de los corchetes solo le ponemos el array booleano
df_sample = df[idx_filtro]
df_sample
```

```python
# ahora colocamos country y year como indices
df_sample.set_index(['Country', 'year'], inplace = True)
df_sample.head()
```

```python
# ordenamos los índices
df_sample = df_sample.sort_index()
df_sample
```

```python
# una forma de recorrer entre los indices
df_sample.loc['Peru',:].loc['2020',:]
```

```python
# otra forma de recorrer entre los indices
df_sample.xs(['Peru', '2020'])
```

```python
df_sample.xs('2020', level='year')
```

```python
# ahora aplicamos eso a todo el dataframe
df_countries = df.set_index(['Country', 'year']).sort_index(ascending = [True, True])
df_countries
```

```python
# para coger pedazos de nuestro dataframe
ids = pd.IndexSlice # esto es una función
# y sobre nuestro dataframe usare 'ids'
df_countries.loc[ids['Aruba': 'Austria', '2018': '2020'],:].sort_index()
```

```python
# para los indices del primer nivel
df_countries.index.get_level_values(0)
```

```python
# para los índices del segundo nivel
df_countries.index.get_level_values(1)
```

```python
# para entrar a un valor en específico
df_countries['pob']['Peru']['2020']
```

```python
# aplicar funciones son una de las ventajas de usar multi index
df_countries.sum(level = 'year')
```

```python
df_sample.unstack('year')
```

```python
df_sample.unstack('year').loc[:,'pob'].loc[:,'2015':'2020']
```

```python
df_sample.unstack('Country')
```

```python
df_sample.unstack('Country').loc['2015':'2020',:]
```
