```python
# para convertir y encontrar el formato más adecuado para las variables
df_meteorites.convert_dtypes().dtypes
# con esto pandas cambia los formatos object por string para agilizar el procesamiento
# de los datos así mismo los enteros
```

```python
# para saber la variedad de los datos
df_meteorites.nunique()
```

```python
# ya que estas son variables categóricas vamos a cambiar al tipo category0
df_meteorites[['nametype', 'fall']] = df_meteorites[['nametype', 'fall']].astype('category')
```

```python
df_meteorites.info()
```

```python
# para saber los valores únicos
df_meteorites['fall'].unique()
```

```python
# para saber la frecuencia de cada valor único en la categoría
df_meteorites['fall'].value_counts()
```

```python
# para sacar la frecuencia que más se repite
df_meteorites['fall'].value_counts().idxmax()
```

```python
# creación de variables dummies
df_dummies_fall = pd.get_dummies(df_meteorites['fall'])
df_dummies_fall.head()
```

```python
# cambiamos el nombre de las variables dummies
df_dummies_fall.rename(columns = {'Fell': 'fall-fell', 'Found': 'fall-found'}, inplace=True)
df_dummies_fall.head()
```

```python
# unimos este dataframe a nuestro dataframe principal
df_meteorites = pd.concat([df_meteorites, df_dummies_fall], axis = 1)
df_meteorites.head()
```

```python
df_meteorites.sample(5)
```

```python
# agregamos una columna de unos a nuestro dataframe
df_meteorites['ones'] = int(np.ones(1)[0])
```

```python
df_meteorites.head()
```

```python
df_meteorites['year']
```

```python
# errors = 'coerce' = permite que cada vez que encuentre un formato que no 
# corresponde al tiempo simplemente genere una variable no numérica
pd.to_datetime(
    df_meteorites['year'],
    errors = 'coerce',
    format = '%m/%d/%Y %H:%M:%S %p'
)
```

```python
df_meteorites['year'] = pd.to_datetime(
    df_meteorites['year'],
    errors = 'coerce',
    format = '%m/%d/%Y %H:%M:%S %p'
)
```

```python
df_meteorites.convert_dtypes().dtypes
```

```python
# renombramos nuestras columnas
df_meteorites.rename(columns = {'mass (g)':'mass'}, inplace = True)
```

```python
df_meteorites.head()
```

```python
# para obtener los nombres de las columnas
list(df_meteorites)
```

```python
# otra forma de saber los nombres de las columnas
df_meteorites.columns
```

```python
df_meteorites.head()
```

```python
# 'ones' = columna
# axis = 1 : eliminará en axis de columna
# inplace = para que el cambio se realice automáticamente
df_meteorites.drop(['ones'], axis = 1, inplace = True)
```

```python
df_meteorites.head()
```

```python
# haremos una copia para no malograr nuestro dataframe principal
# para copiar tenemos que colocar la función copy y el parámetro deep en True
# porque solo así la copia no apuntará al mismo dataframe
df = df_meteorites.copy(deep = True)
df.head()
```

```python
# ahora podemos eliminar filas columnas en nuestra copia
df.drop(columns = ['id'], inplace = True, axis = 1)
```
