```python
def mean_eur2usd(x):
  return np.mean(x)*1.2
```

**apply** = Agregamos una función

**aggregate** = Agregamos una o más de una función

```python
df.groupby(['sex'])[['total_bill', 'prct_tip']].apply(mean_eur2usd)
```

```python
df.groupby(['sex'])[['total_bill', 'prct_tip']].aggregate(mean_eur2usd)
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(mean_eur2usd)
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(np.std)
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].aggregate([np.mean, np.max])
```

```python
dict_agg = {
    'tip': [min, max],
    'total_bill': [np.mean, mean_eur2usd]
}
dict_agg
```

```python
# aggregate = agg
df.groupby(['sex', 'time'])[['total_bill', 'tip']].agg(dict_agg)
```

```python
# filtro todas las facturas cuyo valor sea mayor a 20
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
```

```python
# data frame eliminando las columnas donde el costo promedio de un plato sea menor a 20
df_filtered = df.groupby(['sex', 'time']).filter(f_filter)
df_filtered
```

```python
df_filtered.groupby(['sex', 'time']).count()
```

## Groupby para variables categóricas

```python
df['ones'] = 1
```

```python
df
```

```python
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum()
df_g
```

```python
# level = agrupar en el nivel 0 o nivel género
df_g.groupby(level = 0).apply(
    lambda x:
    x / x.sum() * 100 # a cada valor lo dividimos en la suma del grupo
)
# ahora convertimos las variables numéricas a porcentaje
```

## Convertimos una variable nunérica a categórica

```python
df['bin_total'] = pd.cut(df['total_bill'], bins=[3,18,35,60])
df
```

```python
df.groupby(['time', 'bin_total'])[['ones']].count()
```

```python
df.groupby(['time', 'bin_total'])[['ones']].count().groupby(level = 0).apply(
    lambda x:
    x / x.sum() * 100 # a cada valor lo dividimos en la suma del grupo
)
# ahora convertimos las variables numéricas a porcentaje
```

```python
df_gp = df.groupby(['sex', 'time'])[['total_bill']].mean().reset_index()
df_gp
```

```python
# vemos como influye que un hombre o mujer que fuma gasta y da propina
df.pivot_table(index=['sex', 'smoker'], values=['tip', 'total_bill'], aggfunc=[np.mean, np.median, np.std])
# nos damos con la sorpresa que los hombres fumadores gastan más y dan más propina
```

```python
# ahora quisieramos saber según el sexo qué dias se dá más propina según el sexo
df_sex_smoker_time = df.pivot_table(values = ['tip', 'total_bill'], columns = ['day'], index = ['sex', 'smoker', 'time'], aggfunc=[np.median, max])
df_sex_smoker_time
```

```python
df_sex_smoker_time.fillna(0)
```

```python
df.groupby(['sex', 'smoker', 'time', 'day'])[['tip', 'total_bill']].aggregate([np.median, max]).reset_index().pivot_table(values = ['tip', 'total_bill'], index = ['sex', 'smoker', 'time'], columns = ['day'])
```
